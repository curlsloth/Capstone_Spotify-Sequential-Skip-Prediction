```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

leader_df = pd.read_csv('../data/interim/leaderboard.csv')
leader_df
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(12,4), ncols=2) 


graph1 = sns.histplot(data = leader_df, x = 'AP', bins = 50, kde = True, ax = ax1)
graph1.axvline(0.334, linestyle = '--', c = 'k', label='baseline')
graph1.axvline(0.57, linestyle = '-', c = 'r', label='current model')
ax1.legend( loc ='best')


graph2 = sns.histplot(data = leader_df, x = 'first', bins = 50, kde = True, ax = ax2)
baseline2 = graph2.axvline(0.5, linestyle = '--', c = 'k', label='baseline')
model2 = graph2.axvline(0.744, linestyle = '-', c = 'r', label='current model')
ax2.legend(loc ='best')
ax2.set_ylabel('First prediction accuracy')


fig.suptitle('Leaderboard')


plt.show()
```
