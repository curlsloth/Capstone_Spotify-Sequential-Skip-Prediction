```python
def weight_perSession(df_test_pred):
    import math
    a = df_test_pred['session_position'] - np.floor(df_test_pred['session_length']/2)
    if (a<=0).sum()>0:
        raise Exception('***a number equal or below 0***')
    weight = [1 / x for x in a]
    weight = np.array(weight)
    return weight
```
